Encoding

```python
dataset
```

```python
"""
sex 
Smoker
region

"""
```

```python
Le = LabelEncoder()
dataset["new_sex"] = Le.fit_transform(dataset["sex"])
dataset["new_smoker"] = Le.fit_transform(dataset["smoker"])
dataset["New_region"] = Le.fit_transform(dataset["region"])
```

```python
dataset
```

Target column

```python
x = dataset[["age","new_sex","bmi","children","new_smoker","New_region"]]
y = dataset["charges"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
x.shape, x_train.shape, x_test.shape
```

model training 

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
# prediction on training data
training_data_prediction = lr.predict(x_train)
```

```python
r2_train = metrics.r2_score(y_train, training_data_prediction)*100
print("R squared value",r2_train)
```

```python
# prediction on testing data
testing_data_prediction = lr.predict(x_test)
```

```python
r2_test = metrics.r2_score(y_test, testing_data_prediction)*100
print("R squared value ",r2_test)
```

Building a predictive System

```python
input_data=(19, 0, 27, 0, 1, 3)

input_data_as_numpy = np.asarray(input_data)  # changing input data to a numpy array

input_data_reshape = input_data_as_numpy.reshape(1,-1)  # reshape the array

prediction = lr.predict(input_data_reshape)

print("The insurance cost in USA: ",prediction[0])
```
